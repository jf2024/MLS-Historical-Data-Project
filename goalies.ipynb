```python
# get descriptive statistics for the goalies - slide 5
# and lets make it specific for goalies that played for SJ, delete duplicates if its in the same year
goalies = pd.read_csv('all_goalkeepers.csv')
goalies_sj = goalies[(goalies['Club'].str.strip() == 'SJ') & (goalies['Season'] == 'reg') & (goalies['GP'] >= 5)]
goalies_sj = goalies_sj.drop_duplicates(subset=['Year', 'Player'], keep='first')
goalies_sj

#describe only GAA, Sv%, ShO, W%
goalies_sj = goalies_sj[['GAA', 'Sv%', 'ShO', 'W%']]
goalies_sj.describe()

#get mode and median for GAA, Sv%, ShO, W%
goalies_sj.mode().iloc[0]
goalies_sj.median()

#make a histogram with the goalies_sj based on GAA, SV%, ShO and W%
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# GAA histogram
axs[0, 0].hist(goalies_sj['GAA'], bins=10)
axs[0, 0].set_title('Histogram of GAA for SJ Goalies')
axs[0, 0].set_xlabel('GAA')
axs[0, 0].set_ylabel('Frequency')

# Sv% histogram
axs[0, 1].hist(goalies_sj['Sv%'], bins=10)
axs[0, 1].set_title('Histogram of Save Percentage for SJ Goalies')
axs[0, 1].set_xlabel('Save Percentage')
axs[0, 1].set_ylabel('Frequency')

# ShO histogram
axs[1, 0].hist(goalies_sj['ShO'], bins=10)
axs[1, 0].set_title('Histogram of Clean Sheets for SJ Goalies')
axs[1, 0].set_xlabel('Clean Sheets')
axs[1, 0].set_ylabel('Frequency')

# W% histogram
axs[1, 1].hist(goalies_sj['W%'], bins=10)
axs[1, 1].set_title('Histogram of Win Percentage for SJ Goalies')
axs[1, 1].set_xlabel('Win Percentage')
axs[1, 1].set_ylabel('Frequency')

plt.subplots_adjust(hspace=0.5)
plt.show()
```
